```python
import pandas as pd

Sample_Set = pd.read_csv ("01-01-2021.csv") # https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/01-31-2022.csv
print (Sample_Set)
```

```python
Population = 40000000 # data can be found at https://www.census.gov/quickfacts/CA
Effect = 0.95
False_Positive = 0.42 # data can be found at https://pubmed.ncbi.nlm.nih.gov/34741305/#:~:text=Conclusions%3A%20This%20systematic%20review%20showed,the%20disease%20burden%20among%20the
```

```python
Nset = ['California']
new_df = Sample_Set[Sample_Set['Province_State'].isin(Nset)]
print(new_df)
```

```python
JanCon = Sample_Set.iloc[5]['Confirmed']
print(JanCon)
```

```python
JanConRate = JanCon/Population
NegJanConRate = 1 - JanConRate
print(JanConRate, NegJanConRate)
```

```python
P_Effect = JanConRate * Effect + NegJanConRate * False_Positive
P_Co_Ca_Jan = Effect / P_Effect * JanConRate
print(P_Co_Ca_Jan)
```

```python
a_list = []
a_list.append(P_Co_Ca_Jan)
print(a_list)
```
